```python
# misc
import os
import shutil
import math
import datetime
# plots

# image operation
import cv2

from IPython.display import YouTubeVideo

from __future__ import unicode_literals
import youtube_dl
```

# 1. Select the youtube video by the extension

```python
YOUTUBE_ID = 'U2vzsSmqGg8'

YouTubeVideo(YOUTUBE_ID)
```

# 3. Run the download function

```python
# download the youtube with the given ID
ydl_opts = {

     'outtmpl': '../01 Script Inputs/%(extractor_key)s/%(title)s.%(ext)s',
}
with youtube_dl.YoutubeDL(ydl_opts) as ydl:
    ydl.download(['https://www.youtube.com/watch?v='+YOUTUBE_ID])

#Move the files out of the youtube file 
source_dir = '../01 Script Inputs/Youtube'
target_dir = '../01 Script Inputs'
    
file_names = os.listdir(source_dir)
    
for file_name in file_names:
    shutil.move(os.path.join(source_dir, file_name), os.path.join(target_dir, file_name))

#Delete the youtube file 
os.rmdir('../01 Script Inputs/Youtube')
```
